Having a 1% edge simulation

```python
import numpy as np
import pandas as pd
def generate_returns(n = 100000, edge = 0.01):
    probabillities = [np.random.rand()-0.5 + edge/2 for _ in range(n)]
    returns = [np.log(101/100) if p > 0 else np.log(99/100) for p in probabillities]
    return pd.Series(returns)
generate_returns()
```

```python
logret = generate_returns()
print('Sharpe: ', logret.mean()/logret.std()*100000/(252))
logret.cumsum().plot()
```
